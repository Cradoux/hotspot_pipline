```python
import joypy
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import cm
import numpy as np
```

```python

df = pd.read_csv('validation/scores_validation.csv')
df = df[df['scores'] <25]
df
grouped = df.groupby('pdb', as_index = False).median()
# grouped['lq'] = df.groupby('pdb', as_index = False)['scores'].quantile(.10)
# grouped['uq'] = df.groupby('pdb', as_index = False)['scores'].quantile(.90)
grouped = grouped.rename(index=str, columns={'scores':'median_scores'} )
df = df.merge(grouped, on='pdb')
df = df.sort_values(by=['median_scores'], ascending = [False])
df['colour'] = '#D33C60'
df.loc[(df['tractability'] == 'd'), 'colour'] = '#75D8D5'
```

```python
df['prot_id'] = df['tractability'] + '_' + df['pdb']
df2 = df.groupby('prot_id',sort=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import cycle

from sklearn import svm, datasets
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from scipy import interp

n_classes = 2
y_score = df3['median_scores']
y_test = df3['tractability']

fpr, tpr, thresholds = metrics.roc_curve(y_test, y_score, pos_label='d')
roc_auc = metrics.auc(fpr,tpr)

# Compute ROC curve and ROC area for each class
# fpr = dict()
# tpr = dict()
# roc_auc = dict()
# for i in range(n_classes):
#     fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
#     roc_auc[i] = auc(fpr[i], tpr[i])

# # Compute micro-average ROC curve and ROC area
# fpr["micro"], tpr["micro"], _ = roc_curve(y_test.ravel(), y_score.ravel())
# roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])
```

```python
plt.figure()
lw = 2
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Tractability from Hotspot Score Distribution')
plt.legend(loc="lower right")
plt.show()
```
